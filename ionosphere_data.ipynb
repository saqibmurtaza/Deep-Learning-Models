SPLIT DATA TO TRAIN AND TEST 60:40; USING SK LIB

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.4)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from keras import Sequential,models,layers
```

```python
network = models.Sequential()
network.add(layers.Dense(16,activation='relu',input_shape=(33,)))
```

```python
network.compile(optimizer='rmsprop',loss='binary_crossentropy',metrics='accuracy')
```

```python
history = network.fit(x_train,y_train,
                      batch_size=10,
                      epochs=100)
```

```python
prediction = network.predict(x_test)
```

```python
prediction.shape
```

```python
network.evaluate(x_test,y_test)
```

```python
network = models.Sequential()
network.add(layers.Dense(16,activation='relu',input_shape=(x_train.shape[1],)))
network.add(layers.Dense(8,activation='relu'))
network.add(layers.Dense(1,activation='sigmoid'))
```

```python
history = network.fit(x_train,y_train,batch_size=8,epochs=50)
```
